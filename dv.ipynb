```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import time

def extract_postal_code_and_commune(address_text):
    # Find all sequences of 5 digits in the address
    postal_codes = re.findall(r'\b\d{5}\b', address_text)
    if postal_codes:
        # The last occurrence of a 5-digit sequence is the postal code
        postal_code = postal_codes[-1]
        # The index of the last occurrence of the postal code in the address text
        index = address_text.rfind(postal_code)
        # Extract the commune name which follows the postal code
        commune = address_text[index + 5:].strip()  # +5 to start after the postal code
        return postal_code, commune
    else:
        # Return None if no postal code is found
        return None, None

# Base URL without the page number
base_url = 'https://www.sanitaire-social.com/annuaire-soins-et-aide-a-domicile/services-d-aide-a-domicile-sad/liste-normandie'

# Initialize a list to store addresses
address_data = []

# Loop through the pages
for page in range(1, 13):  # Adjust the range based on the actual number of pages
    # Construct the URL for the current page
    url = f"{base_url}?page={page}"
    
    # Send a GET request to the page
    response = requests.get(url)
    
    # Parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all elements containing the address
    address_elements = soup.find_all('p', class_='card-text address')
    
    # Extract and store the data from each element
    for element in address_elements:
        address_text = element.get_text(strip=True)
        # Extract postal code and commune
        postal_code, commune = extract_postal_code_and_commune(address_text)
        if postal_code and commune:
            address_data.append({'Postal Code': postal_code, 'Commune': commune})
    
    # Be polite: wait a second before scraping the next page
    time.sleep(1)

# Create a DataFrame from the scraped data
df_addresses = pd.DataFrame(address_data)

# Display the first few entries in the DataFrame
print(df_addresses.head())

# Save the DataFrame to a CSV file if needed
# df_addresses.to_csv('normandie_sad_addresses.csv', index=False)
```

```python
# Perform the aggregation
aggregated_by_commune = df_addresses.groupby('Commune').size().reset_index(name='Count')

# Sort the results by the count, in descending order, to see the communes with the most SADs
aggregated_by_commune = aggregated_by_commune.sort_values(by='Count', ascending=False)

# Display the aggregated DataFrame
print(aggregated_by_commune.head())  # Shows the top communes by count
```
